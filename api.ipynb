```python
import requests

URL = "https://api.waterdata.usgs.gov/ogcapi/v0/collections/latest-continuous/items"

# Utah bbox (approx)
bbox_ut = "-114.05,36.99,-109.04,42.00"

params = {
    "f": "json",
    "bbox": bbox_ut,
    "limit": 25,   # start small so you can see structure
}

r = requests.get(URL, params=params, timeout=30)
r.raise_for_status()

data = r.json()
features = data["features"]

print("returned:", len(features))
print("top-level keys:", data.keys())
print("one feature keys:", features[0].keys())
print("properties keys:", features[0]["properties"].keys())
print("example properties:", {k: features[0]["properties"].get(k) for k in
                             ["monitoring_location_id","parameter_code","time","value","unit_of_measure","approval_status"]})
```

```python
import pandas as pd

df = pd.json_normalize(features)
df.head()
```
